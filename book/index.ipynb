## Basics

You can view console output:

````typescript
console.log(
  JSON.stringify(
    {
      hello: 'world'
    },
    null,
    2
  )
);
````

But it's easier to read if you use the convenience `jupyter` object:

````typescript
jupyter.out(
  JSON.stringify(
    {
      hello: 'world'
    },
    null,
    2
  )
);
````

And you can render HTML on your Notebook!

````typescript
jupyter.dom.window.document.body.innerHTML = `
  <img src="https://media.giphy.com/media/VVgRNcBKp64NO/giphy.gif"/>
`;

jupyter.renderDom();
````
