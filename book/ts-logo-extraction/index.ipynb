```typescript
import getPixels from 'get-pixels';

getPixels(
  'typescript-pixel-art.png',
  (err, { data }) => {
    if (err) throw err;

    const argb = Object.values(data);

    const output = [];

    while (argb.length) {
      const pixel = argb.splice(0, 4);

      output.push(pixel[1] === 122 ? 0 : 1);
    }

    jupyter.out(
      JSON.stringify(output)
    );
  }
);
```
